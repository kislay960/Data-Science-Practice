```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```
X,Y = make_blobs(n_samples = 399, centers = 2, n_features = 2, random_state = 2)
#print(X)
```

```
plt.scatter(X[:,0],X[:,1],c=Y)
plt.show()
```

```
#plt.style.use('seaborn-ticks')
x,y=map(int,input('Enter coordinates separated by space\n').split())
query=[x,y]
plt.scatter(X[:,0],X[:,1],c=Y)
plt.scatter(query[0],query[1],color='red')
plt.show()
```

```
def dist(x,y):
  return sum((x-y)**2)
def knn(x,y,query,k=4):
  vals=[]
  for i in range(X.shape[0]):
    distance = dist(query, X[i])
    vals.append((distance,Y[i]))
  vals = sorted(vals)
  vals=vals[:k]
  vals=np.array(vals)
  new_vals = np.unique(vals[:,1],return_counts=True)
  max_freq = new_vals[1].argmax()
  return new_vals[0][max_freq]
```

```
print('Predicted Class :',knn(X,Y,query,2))
```
