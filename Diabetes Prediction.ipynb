```
import numpy as np
import pandas as pd
```

```
x_train = pd.read_csv('https://raw.githubusercontent.com/kislay960/Data-Science-Practice/master/Diabetes_XTrain.csv')
y_train = pd.read_csv('https://raw.githubusercontent.com/kislay960/Data-Science-Practice/master/Diabetes_YTrain.csv')
```

```
x_train.head()
```

```
y_train.head()
```

```
x_train = x_train.values
y_train = y_train.values
```

```
x_test = pd.read_csv('https://raw.githubusercontent.com/kislay960/Data-Science-Practice/master/Diabetes_Xtest.csv')
```

```
x_test = x_test.values
```

```
def dist(x1,x2):
    return np.sqrt(sum((x1-x2)**2))

nearestpoints = 10
def knn(X,Y,test,k=nearestpoints):
    vals = []
    m = X.shape[0]
    for i in range(m):
        d = dist(test,X[i])
        vals.append((d,Y[i]))
    vals = sorted(vals)
    vals = vals[:k]
    vals = np.array(vals)
    new_vals = np.unique(vals[:,1],return_counts=True)
    index = new_vals[1].argmax()
    pred = new_vals[0][index]
    return(pred)
```

```
outcome = []
for i in range(x_test.shape[0]):
    pred = knn(x_train,y_train,x_test[i])
    outcome.append(int(pred))
```

```
outcome = np.asarray(outcome)
```

```
outcome.shape
```

```
df = pd.DataFrame(outcome)
df.columns = ['Outcome']
```

```
df.to_csv('sample_submission.csv',index=False)
```
