```
print(ct.getCleanReview('This was awesome an awesome movie'))
```

```
x = ["This was awesome an awesome movie",
     "Great movie! I liked it a lot",
     "Happy Ending! awesome acting by the hero",
     "loved it! truly great",
     "bad not upto the mark",
     "could have been better",
     "Surely a Disappointing movie"]
y = [1,1,1,1,0,0,0]
```

```
x = ['']
```

```
x_test = ["I was happy & happy and I loved the acting in the movie",
          "The movie I saw was bad"]
```

```
x_clean = [ct.getCleanReview(i) for i in x]
xtest_clean = [ct.getCleanReview(i) for i in x_test]
```

```
print(x_clean)
print(xtest_clean)
```

```
from sklearn.feature_extraction.text import CountVectorizer
```

```
cv = CountVectorizer()
x_vec = cv.fit_transform(x_clean).toarray()
print(x_vec)
```

```
x_vec.shape
```

```
cv.vocabulary_
```

```
xtest_vec = cv.transform(xtest_clean).toarray()
xtest_vec
```

```
from sklearn.naive_bayes import MultinomialNB, BernoulliNB, GaussianNB
```

```
mnb = MultinomialNB()
```

```
mnb.fit(x_vec,y)
```

```
mnb.predict(xtest_vec)
```

```
mnb.predict_proba(xtest_vec)
```

```
bnb = BernoulliNB()
```

```
bnb.fit(x_vec,y)
```

```
bnb.predict(xtest_vec)
```

```
bnb.predict_proba(xtest_vec)
```
