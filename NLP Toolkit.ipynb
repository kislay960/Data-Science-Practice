Bag of words pipeline - 


1.   Get the data from some source or some corpus
2.   tokenisation(breaking document into sentences and sentences into pipeline)
1.   stopword removal(removing prepositions, determiners, articles etc)
2.   stemming - converting all the form of a verb into base form(runs,running = run)
1.   Building vocab - list of words after the above filteration is vocab.
2.   vectorization - converting a sentence to a binary vector of size |v| where v is vocab; 0 where word is not present. 1 where word is present.
1.   Classification - now the model goes to the classifier

stopword removal

```
from nltk.corpus import stopwords
```

```
sw = set(stopwords.words('english'))
```

```
print(sw)
```

```
def remove_stopwords(text,stopwords):
  useful_words = [w for w in text if w not in stopwords]
  return useful_words
```

```
text = "i am not bothered about her very much".split()
useful_text = remove_stopwords(text,sw)
print(useful_text)
```

Vectorization with stopword removal

```
def myTokenizer(document):
  words = regex.tokenize(document.lower())
  words = remove_stopwords(words,stopwords=sw)
  return words
```

```
cv = CountVectorizer(tokenizer=myTokenizer)
```

```
vectorized_corpus = cv.fit_transform(corpus).toarray()
```

Tf-idf Normalisation

*   Avoid features that occur very often, becauase they contain less information
*   Information decreases as the number of occurences increases across different type of documents

*   So we define another term - term-document-frequency which associates a weight with every term

tf*idf value must be high(around 1) for more weight(importance)


idf = log(N/count(t,D)) N is total number of documents and count(t,D) specifies how many t appeared across all documents.

```
sent_1  = "this is good movie"
sent_2 = "this was good movie"
sent_3 = "this is not good movie"
corpus = [sent_1,sent_2,sent_3]
```

```
from sklearn.feature_extraction.text import TfidfVectorizer
```

```
tfidf = TfidfVectorizer()
```

```
vc = tfidf.fit_transform(corpus).toarray()
```

```
print(vc)
```

```
tfidf.vocabulary_
```
