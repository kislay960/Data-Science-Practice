```
import numpy as np
import pandas as pd
```

```
df = pd.read_csv('https://raw.githubusercontent.com/kislay960/Data-Science-Practice/master/mushrooms.csv')
```

```
df.head()
```

```
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```
le = LabelEncoder()
ds = df.apply(le.fit_transform)
ds.head()
```

```
data = ds.values
data_x = data[:,1:]
data_y = data[:,0]
```

```
x_train, x_test, y_train, y_test = train_test_split(data_x, data_y, test_size=0.2)
```

```
x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

```
def prior_probab(y,label):
  total_examples = y.shape[0]
  class_examples = np.sum(y==label)
  return float(class_examples/total_examples)

def conditional_probab(x,y,feature_col,feature_val,label):
  x = x[y==label]
  numerator = np.sum(x[:,feature_col]==feature_val)
  denominator = np.sum(y==label)
  return numerator/float(denominator)

def predict(x,y,xtest):
  classes = np.unique(y)
  no_of_features = x.shape[1]
  post_probs = []
  for label in classes:
    likelihood = 1.0
    for f in range(no_of_features):
      cond = conditional_probab(x,y,f,xtest[f],label)
      likelihood *= cond
    prior = prior_probab(y,label)
    post_prob = likelihood*prior
    post_probs.append(post_prob)
  pred = np.argmax(post_probs)
  return pred
```

```
output = predict(x_train,y_train,x_test[0])
print(output)
print(y_test[0])
```

```
def score(x_train,y_train,x_test,y_test):
  pred = []
  for i in range(x_test.shape[0]):
    pred_label = predict(x_train,y_train,x_test[i])
    pred.append(pred_label)
  pred = np.array(pred)
  accuracy = np.sum(pred==y_test)/y_test.shape[0]
  return accuracy
```

```
score(x_train,y_train,x_test,y_test)
```
