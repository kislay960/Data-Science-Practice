```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
```

```
x = pd.read_csv('https://raw.githubusercontent.com/kislay960/Data-Science-Practice/main/Logistic_X_Train.csv').values
y = pd.read_csv('https://raw.githubusercontent.com/kislay960/Data-Science-Practice/main/Logistic_Y_Train.csv').values
```

```
x_train = np.array(x)
y_train = np.array(y)
```

```
y_train = y_train.reshape((2250,))
```

```
plt.scatter(x_train[:,0],x_train[:,1],c=y_train)
plt.show()
```

```
model = Sequential()
model.add(Dense(256,activation='relu',input_shape=(2,)))
model.add(Dense(128,activation='relu'))
model.add(Dense(64,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
model.summary()
```

```
model.compile(optimizer ='adam',loss = 'binary_crossentropy',metrics=['accuracy'])
```

```
hist = model.fit(x_train,y_train,shuffle=True,batch_size = 100, epochs = 15)
```

```
xtest = pd.read_csv('https://raw.githubusercontent.com/kislay960/Data-Science-Practice/main/Logistic_X_Test_.csv').values
```

```
pred = model.predict(xtest)
```

```
for i in range(pred.shape[0]):
    if pred[i] >=0.5:
        pred[i] =1
    else:
        pred[i] =0
pred = pred.reshape((750,))
print(pred)
```

```
pred=pd.DataFrame(pred,columns=["label"])
pred.to_csv("Classify_points.csv",index=False)
```
