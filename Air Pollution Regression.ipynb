```
import numpy as np
import pandas as pd
```

```
train = pd.read_csv('https://raw.githubusercontent.com/kislay960/Data-Science-Practice/master/Train.csv')
test = pd.read_csv('https://raw.githubusercontent.com/kislay960/Data-Science-Practice/master/Test.csv')
```

```
train = train.values
```

```
xtrain = train[:,:-1]
```

```
xtrain
```

```
ones = np.ones((xtrain.shape[0],1))
```

```
xtrain = np.hstack((ones,xtrain))
xtrain
```

```
ytrain = train[:,-1]
```

```
def hypothesis(x,theta):
  return np.dot(x,theta)

def error(x,y,theta):
  error = 0.0
  n = x.shape[0]
  y_ = hypothesis(x,theta)
  error = np.sum((y-y_)**2)
  return error/n

def gradient(x,y,theta):
  y_ = hypothesis(x,theta)
  grad = np.dot(x.T,(y_ - y))
  m = x.shape[0]
  return grad/m

def gradient_descent(x,y,learning_rate = 0.1,max_iters=300):
  n = x.shape[1]
  theta = np.zeros((n,))
  error_list = []
  for i in range(max_iters):
    e = error(x,y,theta)
    error_list.append(e)
    grad = gradient(x,y,theta)
    theta = theta - learning_rate*grad
  return theta,error_list
```

```
theta,error_list = gradient_descent(xtrain,ytrain)
y_ = []
for i in range(xtrain.shape[0]):
  pred = hypothesis(xtrain[i],theta)
  y_.append(pred)
y_ = np.array(y_)
```

```
def r2_score(y,y_):
  num = np.sum((y-y_)**2)
  denom = np.sum((y-y.mean())**2)
  score = 1 - num/denom
  return score*100
```

```
r2_score(ytrain,y_)
```
